# Now You: Fit a Decision Tree Classifier to the data
---
Let's create our `X` and `y` in the usual (fast, maybe sloppy) way. We'll use every x-variable we can, dummifying when needed.

Remember to binarize `y`!

```python
# Create dummies for the `ChestPain`, `Thal`, and `AHD` columns.
# Be sure to set `drop_first=True`.

X = pd.get_dummies(
    df.drop(columns='AHD'),
    columns=['ChestPain', 'Thal'],
    drop_first=True
)

y = df['AHD'].map({'Yes': 1, 'No': 0})
```

```python
y.mean()
```

```python
# Split data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=42)
```

Our target column will be `AHD_Yes`: 
- 1 means the patient has heart disease
- 0 means they aren't diagnosed with heart disease

```python
# What is the accuracy of our baseline model?
y.value_counts(normalize=True)
```

What does a false positive mean in this case? False negative? Which is worse?

```python
# Let's fit our "dumb" model - an overfit decision tree

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
print(tree.score(X_train, y_train))
print(tree.score(X_test, y_test))
```

# Let's bag it!
---

```python
tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree, n_estimators=100, random_state=2022)
```

```python
bag.fit(X_train, y_train)
```

```python
print(bag.score(X_train, y_train))
print(bag.score(X_test, y_test))
```

# A model of a model of a model
We can even **Grid Search** over our bagged classifier to see if our ensemble benefits from having a "smart" or "dumb" base classifier.

```python
pgrid = {
    'estimator__max_depth': np.append(np.arange(1, 8), None),
    'estimator__min_samples_leaf': np.arange(1, 12, 2)
    
}

tree = DecisionTreeClassifier()
bag = BaggingClassifier(estimator=tree, n_estimators=100, random_state=2022)
gs = GridSearchCV(bag, pgrid, cv=5, n_jobs=4)
```

```python
%%time
# Q: How many total decision trees am I fitting?
gs.fit(X_train, y_train)
```

```python
pgrid
```

```python
# Q: How many total decision trees am I fitting?
# A:

(8*6*100*5) + 100
```

```python
gs.best_params_
```

```python
print(gs.score(X_train, y_train))
print(gs.score(X_test, y_test))
```

```python
results = pd.DataFrame(gs.cv_results_).sort_values('mean_test_score', ascending=False)
results
```

# Estimating Out-of-Sample Error
---

For bagged models, out-of-sample error can be estimated without using **train/test split** or **cross-validation**!

For each tree, the **unused observations** are called "out-of-bag" observations.

```python
tree = DecisionTreeClassifier(max_depth=4, min_samples_leaf=7)
bag = BaggingClassifier(estimator=tree, n_estimators=100, random_state=2022, oob_score=True)
bag.fit(X, y)
```

```python
bag.oob_score_
```

## Estimating Feature Importance

Bagging increases **predictive accuracy** but decreases **model interpretability** because it's no longer possible to visualize the tree to understand the importance of each feature. This is sometimes called a **black box model**.

But we can still obtain an overall summary of **feature importance** from bagged models.

```python
# Compute feature importances of trees within the bagging classifier and take the mean of the results
importances = np.mean([tree.feature_importances_ for tree in bag.estimators_], axis=0)
features = X.columns

# Sort features by importance in descending order
indices = np.argsort(importances)[::-1]

# Plot the feature importances in descending order
plt.figure(figsize=(10,6))
plt.title("Feature Importance (Sorted) - Bagging Classifier")
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.gca().invert_yaxis()  # To have the most important feature at the top
plt.show();
```

# Random Forests
---

Random forest seeks to improve the variance reduction of bagging by **decorrelating the tree** without adding too much variance back.

We do this by adding another source of randomness: before each split, select a random subset of the input variables as candidates for splitting.

Let's try it out!

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
# max_features=4 is best and n_estimators=150 is sufficiently large.
rfclass = RandomForestClassifier(n_estimators=150,
                              max_features=4,
                              oob_score=True,
                              random_state=1)
rfclass.fit(X, y)
```

```python
# Compute feature importances
pd.DataFrame({'feature':X.columns,
              'importance':rfclass.feature_importances_}).sort_values(by='importance', ascending=False)
```

```python
# Compute the out-of-bag accuracy score.
print((rfclass.oob_score_))

# Find the average accuracy score.
scores = cross_val_score(rfclass, X, y, cv=10, scoring='accuracy')
np.mean(scores)
```

# Bonus: Ensembling anything! (Stacking)
You can also ensemble models of any general architecture. Let's try to do this with all the classification models we know so far!

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr = LogisticRegression(penalty=None)
knn = KNeighborsClassifier(n_neighbors=7)
tree = DecisionTreeClassifier(max_depth=4, min_samples_leaf=7)
```

```python
vc = VotingClassifier([
    ('lr', lr),
    ('knn', knn),
    ('tree', tree)
])
```

```python
vc.fit(X_train, y_train)
```

```python
vc.score(X_test, y_test)
```

# Conclusion and Takeaways
---

## Which model is best?

The best classifier for a particular task is task-dependent. In many business cases, interpretability is more important than accuracy. So, decision trees may be preferred. In other cases, accuracy on unseen data might be paramount, in which case random forests would likely be better (since they typically overfit less). 

Remember that every model is a tradeoff between **bias** and **variance**.

Ensemble models attempt to reduce overfitting by reducing variance but increasing bias (as compared to decision trees). By making the model more stable, we necessarily make it fit the training data less accurately. In some cases this is desired (particularly if we start with lots of overfitting), but for more simply structured data a simple decision tree might be best.
